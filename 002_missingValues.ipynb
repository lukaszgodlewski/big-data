```python
import numpy as np
import pandas as pd
import sklearn

sklearn.__version__
```

```python
data = {
    'size': ['XL', 'L', 'M', np.nan, 'M', 'M'],
    'color': ['red', 'green', 'blue', 'green', 'red', 'green'],
    'gender': ['female', 'male', np.nan, 'female', 'female', 'male'],
    'price': [199.0, 89.0, np.nan, 129.0, 79.0, 89.0],
    'weight': [500, 450, 300, np.nan, 410, np.nan],
    'bought': ['yes', 'no', 'yes', 'no', 'yes', 'no']
}

df_raw = pd.DataFrame(data=data)
df_raw
```

```python
df = df_raw.copy()
df.info()
```

Sprawdzenie braków

```python
df.isnull()
```

Użycie narzędzia SimpleImputer do wstawienia brakujących danych. Można wykorzystać różne strategie - średnia, stała, mediana, najczęstsza wartość

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer.fit(df[['weight']])
```

```python
imputer.statistics_
```

```python
imputer.transform(df[['weight']])
```

```python
df['weight'] = imputer.transform(df[['weight']])
df
```

```python
pd.isnull(df['weight'])
```

```python
df[~pd.isnull(df['weight'])]
```

```python
df.fillna(value='brak')
```

```python
df['size']
```

```python
df['size'].fillna(value='L', inplace=True)
df
```
